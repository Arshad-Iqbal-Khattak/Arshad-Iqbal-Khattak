**Abstract Similarity using BERT**

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
file_path = '/kaggle/input/d2cohennew3/Cohennew3binary.xlsx'
dfb = pd.read_excel(file_path, sheet_name='abstract')
dfb.head()
```

```python
import numpy as np
from transformers import BertModel, BertTokenizer

# Load the pretrained BERT model and tokenizer
model_name = 'bert-base-uncased'
tokenizer = BertTokenizer.from_pretrained(model_name)
bert_model = BertModel.from_pretrained(model_name)
```

```python
from transformers import BertTokenizer, BertModel
import torch

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
bert_model = BertModel.from_pretrained('bert-base-uncased')

def calculate_embedding(text):
    encoded_input = tokenizer.encode_plus(text, add_special_tokens=True, truncation=True, max_length=512, padding='max_length', return_tensors='pt')
    input_ids = encoded_input['input_ids']
    attention_mask = encoded_input['attention_mask']
    
    with torch.no_grad():
        outputs = bert_model(input_ids, attention_mask=attention_mask)
        embeddings = outputs.last_hidden_state.mean(dim=1).squeeze().numpy()
    
    return embeddings
```

```python
# Calculate similarity scores
similarity_scoresTb = []

for index, row in dfb.iterrows():
    embedding1b = calculate_embedding(str(row['Citing Abstract']))
    embedding2b = calculate_embedding(str(row['Cited Abstract']))
    similarityTb = np.dot(embedding1b, embedding2b) / (np.linalg.norm(embedding1b) * np.linalg.norm(embedding2b))
    similarity_scoresTb.append(similarityTb)

# Add similarity scores to the DataFrame
dfb['SimilarityTb'] = similarity_scoresTb
dfb['SimilarityTb'] = dfb['SimilarityTb'].apply(lambda x: round(x, 2))
```

**Similarity of title using Bert**

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
file_path = '/kaggle/input/d2cohennew3/Cohennew3binary.xlsx'
dfbt = pd.read_excel(file_path, sheet_name='title')
dfbt.head()
```

```python
import torch
def calculate_embedding(text):
    input_ids = tokenizer.encode(text, add_special_tokens=True)
    input_ids = torch.tensor(input_ids)[None, :]
    with torch.no_grad():
        outputs = bert_model(input_ids)
        embeddings = outputs.last_hidden_state.mean(dim=1).squeeze().numpy()
    return embeddings
```

```python
similarity_scoresTbt = []

for index, row in dfbt.iterrows():
    embedding1bt = calculate_embedding(str(row['Root_Paper_Title']))
    embedding2bt = calculate_embedding(str(row['Cited-by_Paper_Title']))
    similarityTbt = np.dot(embedding1bt, embedding2bt) / (np.linalg.norm(embedding1bt) * np.linalg.norm(embedding2bt))
    similarity_scoresTbt.append(similarityTbt)

# Add similarity scores to the DataFrame
dfbt['SimilarityTbt'] = similarity_scoresTbt
dfbt['SimilarityTbt'] = dfbt['SimilarityTbt'].apply(lambda x: round(x, 2))
```
